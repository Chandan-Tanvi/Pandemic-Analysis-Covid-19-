```python
States=["Andaman&Nicobar Islands","Andhra Pradesh","Arunachal Pradesh", "Assam", "Bihar", "Chandigarh","Chhattisgarh","Dadra and Nagar Haveli","Delhi","Goa","Gujarat", "Haryana", "Himachal Pradesh", "Jammu and Kashmir", "Jharkand", "Karnataka", "Kerala", "Ladakh", "Lakshadweep", "Maharastra","Manipur", "Meghalaya", "Mizoram", "Maharastra", "Nagaland", "Odisha", "Puduchery", "Punjab", "Rajasthan", "Sikkim", "Tamil Nadu", "Telangana", "Tripura", "Uttar Pradesh", "Uttarakhand", "West Bengal"]

TotalCases=[100,15252,195,8582,10249,446,2940,215,89802,1387,33232,14941,979,7695,2521,16514,4593,990,0,180298,1260,52,160,13861,459,7316,714,5668,18312,101,94049,17357,1396,24056,2947,19170]

Discharged=[50,6988,66,5851,7946,367,2303,86,59992,670,24030,10499,614,4856,1931,8063,2439,694,0,93154,579,42,123,10655,168,5353,272,3867,14574,53,52926,8082,1093,16629,2317,12528]

Deceased=[0,193,1,12,70,6,14,0,2803,4,1867,240,10,105,15,253,24,1,0,8053,0,1,0,581,0,25,12,149,421,0,1264,267,1,718,41,683]

print("Number of States:",len(States))
print("Number of States where cases are active:", len(TotalCases)) #Checking if the number of elements in all lists are equal 
print("Number of States where patients are discharged:", len(Discharged))
print("Number of enteries in the Deceased list:", len(Deceased))

print("\nTotal number of cases in India; Statewise(2nd July 2020):", sum(TotalCases)) 
for j in range(0,len(TotalCases)):
    print("\n", States[j], "-", TotalCases[j])

```

```python
##Number of infected people = [Total cases - (Discharged+Deceased Cases)]

#Finding number of discharged people - 

TotalSick=[] #Declaring a list to calculate number of people no longer affected by Covid
for i in range(0, len(Discharged)):
    TotalSick.append(Discharged[i] + Deceased[i]) #Adding elements to the list using for loop
    
print ("\nNumber of people discarged and deceased due to Covid-19; Stateswise:", sum(TotalSick))
for y in range(0,len(TotalSick)):
    print("\n", States[y], "-", TotalSick[y])
```

```python
#Declaring a list to calculate infected cases 
Infected = []
for x in range(0, len(TotalCases)):
    Infected.append(TotalCases[x]-TotalSick[x])

print("\nNumber of people infected; Statewise:", sum(Infected))
for n in range(0,len(Infected)):
    print("\n",States[n], "-", Infected[n]) 
```
